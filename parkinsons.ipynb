## Splitting Feature and target Variable

```python
X = parkinsons_data.drop(['status'], axis=1)
y = parkinsons_data['status']
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.3, stratify=y, random_state=0)
```

```python
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

## Feature Selection

```python
# Correlation 
plt.figure(figsize=(16,10))
corr = x_train.corr()
sns.heatmap(corr, annot=True, cmap=plt.cm.CMRmap_r)
```

```python
# Function for select highly correlated features 

def correlation(dataset, threshold):
    col_corr = set()
    corr_matrix = dataset.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i][j]) > threshold:
                colname = corr_matrix.columns[i]
                col_corr.add(colname)
    return col_corr
```

```python
corr_features = correlation(x_train, 0.70)
len(set(corr_features))
```

```python
corr_features
```

```python
x_train.drop(corr_features, axis=1, inplace=True)
x_test.drop(corr_features, axis=1, inplace=True)
```

```python
x_train.shape, x_test.shape
```

# Model Building

## 1.) Logistic Regression

```python
from sklearn.linear_model import LogisticRegression

lgr_model = LogisticRegression()
```

```python
lgr_model.fit(x_train, y_train)
```

### Logistic Regression Model Evaluation

```python
lgr_y_pred = lgr_model.predict(x_test)
lgr_y_pred
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix

lgr_accuracy = accuracy_score(y_test, lgr_y_pred)
lgr_precision = precision_score(y_test, lgr_y_pred)
lgr_recall = recall_score(y_test, lgr_y_pred)
lgr_f1score = f1_score(y_test, lgr_y_pred)
```

```python
confusion_matrix(y_test, lgr_y_pred)
```

```python
print('Accuracy : {:.2f}%'.format(lgr_accuracy*100))
print('Precision : {:.2f}%'.format(lgr_precision*100))
print('Recall : {:.2f}%'.format(lgr_recall*100))
print('F1 score : {:.2f}%'.format(lgr_f1score*100))
```

## 2.) Support Vecotor Classifier

```python
from sklearn.svm import SVC

svm_classifier = SVC(kernel='rbf')
```

```python
svm_classifier.fit(x_train, y_train)
```

```python
svm_y_pred = svm_classifier.predict(x_test)
svm_y_pred
```

```python
svm_accuracy = accuracy_score(y_test, svm_y_pred)
svm_precision = precision_score(y_test, svm_y_pred)
svm_recall = recall_score(y_test, svm_y_pred)
svm_f1score = f1_score(y_test, svm_y_pred)
```

```python
confusion_matrix(y_test, svm_y_pred)
```

```python
print('Accuracy : {:.2f}%'.format(svm_accuracy*100))
print('Precision : {:.2f}%'.format(svm_precision*100))
print('Recall : {:.2f}%'.format(svm_recall*100))
print('F1 Score : {:.2f}%'.format(svm_f1score*100))
```
